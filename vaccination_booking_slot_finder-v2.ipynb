```python
#create first column dataframe
centre_list = ['Vesailles','Velizy','St-Cyr']
df = pd.DataFrame(centre_list, columns=['centre'])
```

```python
options = webdriver.ChromeOptions()
#options.add_argument('headless')

driver = webdriver.Chrome('./chromedriver')
wait_sec=0.5
url_booking = None

for trial in range(1,96):
    url_booking = None
    
    for i,url in enumerate(doctolib_list):

        try:
            driver.get(url)
            time.sleep(wait_sec)
            select_fr = Select(driver.find_element_by_xpath("//select[@id='booking_motive']"))
            select_fr.select_by_visible_text('1ère injection vaccin COVID-19 (Pfizer-BioNTech)')
            time.sleep(wait_sec)
            answer = driver.find_element_by_xpath("//*[@id='booking-content']/div[2]/div[3]/div/div[2]/div/div/div/div[1]/div/div/span/span").text
            
            if answer != "Aucun rendez-vous n'est disponible pour le moment mais de nombreux créneaux vont être mis en ligne dans les jours à venir. Vous pouvez également prendre rendez-vous dans un autre lieu de vaccination.":
                url_booking = url
                break
        except NoSuchElementException:
            url_booking = url
            break
    
        

    if url_booking is not None:
        print('URL FOR BOOKING APPOINTMENT  : ',  url_booking)
        now = datetime.now()
        print(now)
        break

    now = datetime.now()
    current_time = now.strftime("%m/%d-%H:%M:%S")
    print('trial n.', str(trial), 'at', str(current_time) , 'terminé. Sleep for 3 min before next round')
    time.sleep(60)

Audio(sound_file,rate = 22050, autoplay=True)
```
